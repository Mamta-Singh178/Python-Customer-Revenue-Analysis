```python
#5. Visualise to show the percentage of active and inactive customers--pie chart--groupby--value_counts
# Calculate the number of active and inactive customers

status_count = customers_df['Status'].value_counts()

# Create the pie chart
# First define the dimension of chart

plt.figure(figsize=(10,6))
plt.pie(status_count,labels=status_count.index,autopct = '%1.1f%%',colors=['#52A736','#D2042D'])
plt.title("Active vs Inactive Customers")
plt.show
```
